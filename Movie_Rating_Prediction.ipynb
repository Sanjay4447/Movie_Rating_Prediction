```python
import pandas as pd
```

```python
data = pd.read_csv("C:\\Users\\sanja\\Downloads\\Movie dataset.csv")
data.head()
```

```python
#IMPORTING MODELS AND TECHNIQUES USING SCIKIT-LEARN
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression
```

```python
data = data[1:]
data.head()
```

```python
#DROPPED ROWS BASED ON RATING VALUE(i.e,, NULL)
```

```python
df = data.drop(index=1)
df = pd.DataFrame(data)
df = df.dropna(subset=['Rating'])
```

```python
#PERFORMED ONE HOT ENCODING FOR CATEGORICAL VARIABLES TO CONVERT THEM INTO NUMERICAL VECTORS
```

```python
genre_dummies = df['Genre'].str.get_dummies(sep=', ')
df = pd.concat([df, genre_dummies], axis=1)
df = df.drop(columns=['Genre'])
dir_dummies = df['Director'].str.get_dummies(sep=', ')
df = pd.concat([df, dir_dummies], axis=1)
df = df.drop(columns=['Director'])
newact = df['Actor 1'].astype(str) + '|' + df['Actor 2'].astype(str) + '|' + df['Actor 3'].astype(str)
newactdum = newact.str.get_dummies(sep='|')
df = pd.concat([df, newactdum], axis=1)
df = df.drop(columns=['Actor 1', 'Actor 2', 'Actor 3'])
df.head()
```

```python
#MODEL TRAINING AND TESTING UAING ALOGORITHMS
```

```python
x = df.drop(columns=['Rating','Name','Year','Duration','Votes'])
y = df["Rating"]
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.4,random_state = 42)
model1 = RandomForestRegressor()
model1.fit(x_train,y_train)
```

```python
pred_y = model1.predict(x_test)
```

```python
#PERFORMANCE USING RANDOMFOREST REGRESSOR
```

```python
print(mean_squared_error(y_test,pred_y))
print(r2_score(y_test,pred_y))
```

```python
model2 = LinearRegression()
model2.fit(x_train,y_train)
```

```python
pred_y1 = model2.predict(x_test)
```

```python
#PERFORMANCE USING LINEAR REGRESSION
```

```python
print(mean_squared_error(y_test,pred_y1))
print(r2_score(y_test,pred_y1))
```

```python
model3 = DecisionTreeRegressor()
model3.fit(x_train,y_train)
```

```python
pred_y3 = model3.predict(x_test)
```

```python
#PERFORMANCE USING DE
```

```python
print(mean_squared_error(y_test,pred_y3))
print(r2_score(y_test,pred_y3))
```
